```python
import pandas as pd
```

```python
vine_review_data = "vine_table.csv"
```

```python
vine_df = pd.read_csv(vine_review_data)
```

```python
vine_df
```

```python
#Step 1
vine_reviews_df = vine_df[(vine_df["total_votes"] > 20)]
vine_reviews_df
```

```python
#Step 2
vine_2_df = vine_reviews_df[((vine_reviews_df["helpful_votes"] / vine_reviews_df["total_votes"])  >= 0.5 )]
vine_2_df
```

```python
#Step 3 
vine_3_df = vine_2_df[(vine_2_df["verified_purchase"] == "Y")]
vine_3_df
```

```python
#Step 4
vine_4_df = vine_2_df[(vine_2_df["verified_purchase"] == "N")]
vine_4_df
```

```python
#Step 5

review_count = vine_df["review_id"].count()
five_star_count = (vine_df["star_rating"] == 5).count()
five_star_countp = (vine_3_df["star_rating"] == 5).count()
five_star_countup = (vine_4_df["star_rating"] == 5).count()

print(review_count)
print(five_star_count)
print(five_star_countp)
print(five_star_countup)
```

```python
#Step 5.1 Percentage of 5 star from two types 
paid_percentage = (float(five_star_countp) / float(five_star_count)) * 100
unpaid_percentage = (float(five_star_countup) / float(five_star_count)) * 100

print(paid_percentage)
print(unpaid_percentage)
```
